```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create the outer figure
fig = make_subplots(rows=4, cols=1)


# Create the inner figure
inner_fig = go.Figure()

# Add a bar chart to the inner figure
inner_fig.add_trace(go.Bar(x=[1, 2, 3], y=[3, 2, 1], name='Inner'))

# Add the inner figure to the outer figure as a subplot
fig.add_trace(inner_fig, row=4, col=1)

fig.update_layout(title='Outer Figure with Inner Figure')
fig.show()
```
